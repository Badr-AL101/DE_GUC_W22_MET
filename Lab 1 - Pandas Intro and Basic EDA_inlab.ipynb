## What is EDA?

Exploratory Data Analysis refers to the critical process of performing initial investigations on data so as to discover patterns,to spot anomalies,to test hypothesis and to check assumptions with the help of summary statistics and graphical representations.

### Data Exploration

Data Exploration is an integral part of EDA (Exploratory Data Analysis). Whatever you decide to do in the later phases (creating/selecting a machine learning model or summarizing your findings), will depend on the assumptions you make in the exploration phase. It’s not a single step phase, but we get to determine a lot about our data during data exploration e.g. checking data distribution, finding correlation, finding outliers and missing values, etc

### Data Visualization

Data Visualizations aren’t part of any specific phase in a data analytics project. We can use visuals to represent the data at any point in our project. We need visualization as a visual summary of the data, because it’s easier to understand for identifying relations and patterns. Many visuals are used in the data exploration phase to find outliers, correlation between features, etc. 

Some examples of the basic and widely used charts: 

*   Line chart
*   Bar chart
-  Histogram
-  Box plot
-  Scatter plot
-  Heatmap

## Getting started - Series
In Pandas the two main data structures are series and dataframes. 

Series is a one-dimensional array holding data of any type. You could think of it as a column in a table.

Firstly, we will import pandas and give it an abbreviation pd for ease of future use.


```python
import pandas as pd
```

Example of a series - To create a series we use the Series method and pass in a list. 

```python
dummy = [5,4,2]
dummy = pd.Series(dummy)
print(dummy)
print(dummy.index,dummy.values)
```

```python
# Access series just like you would access a list, print second item
print(dummy[1])
```

```python
type(dummy),type(dummy[0])
```

```python
# get length of a series, just like a list
len(dummy)
```

As mentioned earlier Pandas library is built on top of the numpy library. This could be seen in the example above as the datatype of the elements inside the series are numpy integers.  

A series could also be indexed by adding the index into the Series method parameter.This would allow us to access the series with our own indexes instead of the conventional numbers

```python
dummy2 = [5,4,2]
dummy2 = pd.Series(dummy2,index = ['x','y','z'])
print(dummy2['x'])
```

```python
print(dummy2)
```

## Dataframes - 
A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.
The best way to think about dataframes is as if they are dictionaries. Where columns represents the keys and the value in inside this column is the value.

You can create a dataframe by using the dataframe method and passing in a 2 dimensional data structure (such as a dictionary).

It is best that if you want to create a dataframe you do so by passing in a dictionary.As this is the best representation of a dataframe with rows and columns (keys are column names and values represents the rows).

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40,70]
}

df = pd.DataFrame(data)

print(df) 
```

## Accessing dataframes

Acessing columns - 

You can access columns just like you would in a dictionary. As mentioned earlier the column represents the key in a dictionary, thus you can just pass the column name inside clode brackets.The type is a pandas series as it represents a 1d column.

```python
## access a column
cals = df['calories']
print(cals)
```

```python
#alternatively 
cals = df.calories
print(cals)
```

```python
type(cals)
```

Accessing rows - using loc or iloc. loc allows you to access the row by passing in its index. 

As we seen earlier the index of each row was assigned automitcally becauase we did not assign an index.
Loc returns a series.

```python
## Print third row
print(df.loc[2])
print(type(df.loc[2]))
```

Another way to access rows is using iloc. This allows you to acces rows by passing in their actual position the dataframe and not their index. I.e if you want to print the 3rd row you would pass in 2, however if you want print the last row you can pass in -1 just like you would in a list. This is not possible with loc because there is no index '-1'. 

```python
## Print third row
print(df.iloc[2])
print(type(df.iloc[2]))
```

```python
## Print last row
print(df.iloc[2])
print(type(df.iloc[2]))
```

```python
# print all rows using slicing
print(df.iloc[:])
```

```python
## Print last row using loc and indexing with -ve number,
# will return an error
print(df.loc[-1])
print(type(df.loc[-1]))
```

A dataframe could also be indexed similar to series.

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data,index = ['day1','day2','day3'])

print(df) 
```

```python
df.loc['day2'],df.iloc[2]
```
